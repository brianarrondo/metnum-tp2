## Test 3: Comparacion de accuracy y time

```python
data = pd.read_csv('tests_csv/test3.csv')

data['K'] = data['K'].apply(lambda K: int(42000 * (K - 1) / K))

data = data.rename(columns={'K': 'Training Amount'})

data
```

```python
sns.lineplot(x='Training Amount', y='Accuracy', hue='Method', data=data)
plt.title("Test 3 - Accuracy")
```

```python
sns.lineplot(x='Training Amount', y='Fit Time', hue='Method', data=data)
plt.title("Test 3 - Fit Time")
```

```python
sns.lineplot(x='Training Amount', y='Prediction Time', hue='Method', data=data)
plt.title("Test 3 - Prediction Time")
```
