# Experimentación - Gráficos

```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Test 1: Knn variando la cantidad de vecinos

```python
data = pd.read_csv('tests_csv/test1.csv')

data
```

```python
sns.lineplot(x='Neighbors', y='Accuracy', hue='Method', data=data)
plt.title("Test 1 - Accuracy")
```

```python
dataTotal = data.drop(columns='Pred time').rename(columns={'Total time': 'Time', 'Method': 'Time type'})
dataTotal['Time type'] = 'Total'

dataPred = data.drop(columns='Total time').rename(columns={'Pred time': 'Time', 'Method': 'Time type'})
dataPred['Time type'] = 'Prediction'

dataTime = pd.concat([dataTotal, dataPred])
```

```python
sns.lineplot(x='Neighbors', y='Time', hue='Time type', data=dataTime)
plt.title("Test 1 - KNN Time")
```

## Test 2: PCA + KNN variando la cantidad de componentes principales

```python
data = pd.read_csv('tests_csv/test2.csv')

data
```

```python
sns.lineplot(x='Components', y='Accuracy', hue='Method', data=data)
plt.title("Test 2 - Accuracy")
```

```python
dataTotal = data.drop(columns='Pred time').rename(columns={'Total time': 'Time', 'Method': 'Time type'})
dataTotal['Time type'] = 'Total'

dataPred = data.drop(columns='Total time').rename(columns={'Pred time': 'Time', 'Method': 'Time type'})
dataPred['Time type'] = 'Prediction'

dataTime = pd.concat([dataTotal, dataPred])
```

```python
grid = sns.lineplot(x='Components', y='Time', hue='Time type', data=dataTime)
grid.set(yscale="log")
plt.title("Test 2 - PCA + KNN Time")
```
